```python
scanner = pysces.Scanner(rc.C.mod)
user_output = [rc.C.fixed]+['ecR1_H','ecR1_E','ecR6_E','ecR6_G','ecR7_G','ecR8_H','ecR7_H','ecR2_B','ecR4_D']
scanner.quietRun = True
scanner.addScanParameter(rc.C.fixed,
                         rc.C.scan_min,
                         rc.C.scan_max,
                         rc.C.scan_points,
                         log=True)
scanner.addUserOutput(*user_output)
scanner.Run()
ec_pat_plt = psctb.utils.plotting.Data2D(rc.C.mod,user_output,scanner.UserOutputResults,rc._ltxe).plot()
ec_pat_plt.ax.set_xlim([rc.C.scan_min,rc.C.scan_max])
ec_pat_plt.ax.semilogx()
ec_pat_plt.ax.set_ylabel('Coefficient Value')
ec_pat_plt.interact()
```
